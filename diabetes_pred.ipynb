```python
import pandas as pd
```

```python
diabetes_data = pd.read_csv('diabetes.csv')
```

```python
X = diabetes_data.drop(columns='Outcome', axis=1)#axis specifies 1 for column and 0 for row
y = diabetes_data['Outcome']
#iloc can also be used or loc 
```

DATA PREPROCESSING 

```python
#DATA PREPROCESSING BY STANDARDIZATION
from sklearn.preprocessing import StandardScaler
```

```python
data_scaler = StandardScaler() 
```

```python
# X_train_scaled = data_scaler.transform(X_train)
# X_test_scaled = data_scaler.transform(X_test)
# print(X)
```

SPLIT THE DATA INTO TRAINING AND TESTING DATA

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, stratify=y ,random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

NOW SCALING THE DATA... NEVER SCALE BEFORE SPLITTING!!!! this is the correction below. The above is the mistake

```python
data_scaler.fit(X_train)
```

```python
X_train_scaled = data_scaler.transform(X_train)
X_test_scaled = data_scaler.transform(X_test)
print(X_train_scaled)
```

TRAINING THE DATA WITH THE SVM MODEL

```python
from sklearn import svm
```

```python
classifier = svm.SVC(kernel='linear')
```

FIT THE TRAINING DATA

```python
classifier.fit(X_train,y_train)
```

```python
X_test_prediction = classifier.predict(X_test) # predict using the test data
X_train_prediction = classifier.predict(X_train) # predict using the train data
```

TRAINING MODEL WITH LOGISTIC REGRESSION MODEL

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
X_test_prediction2 = model.predict(X_test) # predict using the test data
X_train_prediction2 = model.predict(X_train) # predict using the train data 
```

TEST FOR ACCURACY: ACCURACY SCORE, CLASSIFICATION REPORT AND CONFUSION MATRIX

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
print(accuracy_score(X_test_prediction, y_test)) #to check the accuracy of the test prediction
print(accuracy_score(X_train_prediction, y_train))#to check the accuracy of the train prediction



#LOGISTIC REGRESSION MODEL EVALUATION
print(accuracy_score(X_test_prediction2, y_test)) #to check the accuracy of the test prediction
print(accuracy_score(X_train_prediction2, y_train))#to check the accuracy of the train prediction



#FROM THE RESULTS, SVM MODEL PERFORMS SLIGHTLY BETTER THAN LOGISTIC REGRESSION MODEL
```

```python
print(confusion_matrix(X_test_prediction, y_test))
print(confusion_matrix(X_train_prediction, y_train))
```

```python
print(classification_report(X_test_prediction, y_test))
print(classification_report(X_train_prediction, y_train))
```

MAKING A PREDICTIVE SYSTEM

```python
import numpy as np
```

```python
input_data = (5,116,74,0,0,25.6,0.201,30)#creating a sample tuple

#CHANGING THE INPUT DATA TO A NUMPY ARRAY
inputDataAsNumpyArray = np.asarray(input_data)

#RESHAPING THE ARRAY SINCE WE ARE PREDICTING FOR ONLY ONE INSTANCE, UNLIKE THE TEST AND TRAIN CASES
inputDataReshaped = inputDataAsNumpyArray.reshape(1,-1)

#STANDARDIZE THE INPUT DATA
std_data = data_scaler.transform(inputDataReshaped)
print(std_data)


prediction = classifier.predict(std_data)
print(prediction)

if prediction == 0:
    print('The patient is non-diabetic')
else:
    print('The patient is diabetic')
```
